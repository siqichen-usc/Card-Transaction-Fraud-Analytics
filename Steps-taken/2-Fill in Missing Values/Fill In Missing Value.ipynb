```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data=pd.read_excel('card transactions.xlsx')
```

```python
#check number of records
#Merchnum missing, Merch state, Merch zip
data.count()
```

# Fill in State

```python
mydata=data
```

```python
mydata['Merch state'].value_counts()
```

```python
#First fill in "Merch state" with the most common state from the entire dataset
mydata["Merch state"]=mydata["Merch state"].fillna(mydata["Merch state"].mode().iloc[0])
```

```python
mydata['Merch state'].value_counts()
```

```python
#Check if Merch state contains null
mydata.count()
```

# Fill in Merch zip

```python
#Check most frequent value of Merch zip
mydata["Merch zip"].value_counts()
```

```python
mydata[mydata['Merch zip']==38118]
```

```python
#create lookup table for most common Merch zip aggregated by Merch state
state_zip=mydata.groupby("Merch state")["Merch zip"].apply(lambda x:x.mode())\
    .reset_index()
state_zip=state_zip[state_zip.level_1==0].iloc[:,[0,2]]
state_zip.columns=["Merch state","zip"]
```

```python
state_zip
```

```python
#merge with original dataset
merged=mydata.merge(state_zip,on="Merch state",how="left")
```

```python
#fill in na with zip from loockup table
#Fill in 38118 if still null (mainly for the state in numeric format)
merged["Merch zip"]=merged["Merch zip"].fillna(merged['zip']).fillna(38118)
```

```python
#check how many are null
merged["Merch zip"].isnull().sum()
```

```python
merged
```

# Fill in Merchnum

## Group by Merch zip

```python
#Check original numer of null value for Merchnum, 3375 
mydata["Merchnum"].isnull().sum()
```

```python
#create lookup table for most common Merchnum aggregated by Merch zip
zip_Merchnum=merged.groupby("Merch zip")["Merchnum"].apply(lambda x:x.mode())\
    .reset_index()
zip_Merchnum=zip_Merchnum[zip_Merchnum.level_1==0].iloc[:,[0,2]]
zip_Merchnum.columns=["Merch zip","num"]
```

```python
#merge with original dataset
merged=merged.merge(zip_Merchnum,on="Merch zip",how="left")
```

```python
#fill in na with Merchnum from loockup table
merged["Merchnum"]=merged["Merchnum"].fillna(merged['num'])
```

```python
#check how many are null
merged["Merchnum"].isnull().sum()
#still have 235 null value
```

## Group by Merch state

```python
#create lookup table for most common Merchnum aggregated by Merch state
state_Merchnum=merged.groupby("Merch state")["Merchnum"].apply(lambda x:x.mode())\
    .reset_index()
state_Merchnum=state_Merchnum[state_Merchnum.level_1==0].iloc[:,[0,2]]
state_Merchnum.columns=["Merch state","num2"]
```

```python
#merge with original dataset
merged=merged.merge(state_Merchnum,on="Merch state",how="left")
```

```python
#fill in na with Merchnum from loockup table
merged["Merchnum"]=merged["Merchnum"].fillna(merged['num2'])
```

```python
merged["Merchnum"].isnull().sum()
```

```python
merged["Merchnum"].value_counts()
```
